```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Definieer de straal en het middelpunt van de bol
r = 3 * np.sqrt(2)  # straal
center = (3, 0, 3)  # middelpunt (x0, y0, z0)

# Maak een raster voor de plot
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

# Bepaal de coördinaten van de bol
X = center[0] + r * np.outer(np.cos(u), np.sin(v))
Y = center[1] + r * np.outer(np.sin(u), np.sin(v))
Z = center[2] + r * np.outer(np.ones(np.size(u)), np.cos(v))

# Definieer de hoekpunten van de kubus
hoekpunten = np.array([
    [0, 0, 0],  # A
    [6, 0, 0],  # B
    [6, 0, 6],  # C
    [0, 0, 6],  # D
    [0, 6, 0],  # E
    [6, 6, 0],  # F
    [6, 6, 6],  # G
    [0, 6, 6]   # H
])

# Definieer de vlakken van de kubus
vlakken = [
    [hoekpunten[0], hoekpunten[1], hoekpunten[2], hoekpunten[3]],  # Onderkant ABCD
    [hoekpunten[4], hoekpunten[5], hoekpunten[6], hoekpunten[7]],  # Bovenkant EFGH
    [hoekpunten[0], hoekpunten[1], hoekpunten[5], hoekpunten[4]],  # Zijkant ABEF
    [hoekpunten[1], hoekpunten[2], hoekpunten[6], hoekpunten[5]],  # Zijkant BCFG
    [hoekpunten[2], hoekpunten[3], hoekpunten[7], hoekpunten[6]],  # Zijkant CDGH
    [hoekpunten[3], hoekpunten[0], hoekpunten[4], hoekpunten[7]]   # Zijkant DAEH
]

# Maak de gecombineerde plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot de bol
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.6)

# Teken de kubus
ax.add_collection3d(Poly3DCollection(vlakken, facecolors='cyan', linewidths=1, edgecolors='r', alpha=0.5))

# Stel de limieten in voor een groter assenstelsel
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])

# Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('Bol en Kubus in 3D')

# Toon de plot
plt.show()
# in deze figuur ligt het bovenvlak in het z-vlak
```
