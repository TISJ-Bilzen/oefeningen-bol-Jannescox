```python
# geg) bol: x**2+Y**2+Z**2-2x-2y-2z-22 en rechte: (x=1+r, y=r, z=4+r)
#bolvergelijking (x**2+y**2+z**2-2x-2y-2z-22) -> (x-1)**2 + (y-1)**2 +(z-1)**2 - 25 =0
# x waarde van de rechte in de bolvergelijking invullen -> (1+r-1)**2 + (r-1)**2 + (4+r-1)**2 = 25 -> 3(r)**2 + 4r - 15 = 0
# dan discriminant van verglijking nemen: 

EQ1 = Eq(3*(r)**2 + 4*r - 15,0)
SOL1 = solve(EQ1,r)
EQ1, SOL1
# deze r waarde in de rechte invullen en zo de twee snijpunten bepalen: s1(8/3,5/3,17/3), s2(-2,-3,1)
#nu grafische voorstelling: 
```

```python
import numpy as np
import matplotlib.pyplot as plt

#Definieer de straal en het middelpunt van de bol
r = 5 # straal (sqrt(81))
center = (-1, -1, -1) # middelpunt (x0, y0, z0)

#Maak een raster voor de plot
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

#Bepaal de coördinaten van de bol
X = center[-1] + r * np.outer(np.cos(u), np.sin(v))
Y = center[-1] + r * np.outer(np.sin(u), np.sin(v))
Z = center[-1] + r * np.outer(np.ones(np.size(u)), np.cos(v))

#Plot de bol
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.6)

#Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('Bol met middelpunt (-1, -1, -1) en straal 5')

#Toon de plot

import matplotlib.pyplot as plt
x1, y1 = [8/3,5/3,17/3], [-2, -3,1]

plt.plot(x1, y1, marker = 'o')





plt.show()

```
